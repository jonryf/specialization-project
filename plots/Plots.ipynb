## Insert test YCSB

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import seaborn as sns

sns.set_theme(style= "white", palette="spring")



df = pd.DataFrame([['B+-tree','1',300],['B+-tree','2',300],['B+-tree','4',300],['B+-tree','8',300],
                   ['B+-tree with locks','1',330],['B+-tree with locks','2',540],
                   ['B+-tree with locks','4',560],['B+-tree with locks','8',490],
                   ['B+-tree with queue','1',480],['B+-tree with queue','2',500],
                   ['B+-tree with queue','4',565],['B+-tree with queue','8',550]],
                  columns=['group','Threads','val'])




df.pivot("Threads", "group", "val").plot(kind='bar', figsize=(12,4), fontsize=16)

plt.ylabel("Throughput (KOps/sec)", fontdict={'fontsize':16})
plt.xlabel("Threads", fontdict={'fontsize':16})


axes = plt.gca()
axes.set_ylim([0,900])
plt.rcParams.update({'font.size': 40})

fig = plt.gcf()

#fig.set_size_inches(12, 4)
#plt.figure(num=None, figsize=(8, 1), dpi=80, facecolor='w', edgecolor='k') 
fig.savefig('yinsert.png', dpi=800)

# figure(num=None, figsize=(10, 6), dpi=80, facecolor='w', edgecolor='k')

#plt.show()
```
